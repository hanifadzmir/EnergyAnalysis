A utility company invested 50 million dollars to build a new generator that can generate 100 GWh of electricity per year to serve 1,000 consumers. The generator has a fixed production cost of $50/MWh and has a projected lifespan of 20 years. Apart from the generator, the utility also has a fixed 10 million dollars maintainance cost corresponding to this generator and the associated distribution grids to ensure service reliability. Assuming all electricity generated by this new generator will be delivered to consumers in this utility, and the utility has to recover all its investment and operation cost from the electricity retail.

Note: 1GWh = 1,000 MWh = 1,000,000 kWh


1) Calculate the annualized investment cost of this generator assuming an interest rate of 5% per year and a investment period of 20 years. (10%)

Use the following equation for calculating annual payments:
$$Y = P \frac{ i(1 + i)^n }{ (1 + i)^n – 1} $$
where 
- $Y$ is the annual payments
- $P$ is the total amount of loan
- $i$ is the annual interest rate 
- $n$ is the number of years to pay off

- Total amount of loan, $P$ (Investment Cost) = $50,000,000
- Annual interest rate, $i$ = 5%
- Investment period, $n$ = 20 years
$$Y = P \frac{ i(1 + i)^n }{ (1 + i)^n – 1} $$
$$Y = 50,000,000 \frac{ 0.05(1 + 0.05)^{20} }{ (1 + 0.05)^{20} – 1} $$
<mark> Y = 4,012,129.36 

```python
P = 50000000.00
i = 0.05
n= 20.00
Y = ((P)*i*(1+i)**(n))/((1+i)**(n)-1)
Y
```
